```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = pd.read_excel(r'C:\\Users\Trenkin.Sergey\Desktop\Копия Расчет сезонности.xlsx', sheet_name='Сезон по городам').dropna(axis=1, how='all')
```

```python
data
```

```python
scaler = StandardScaler()
scaler.fit(data.drop('Город', axis=1))
```

```python
scaled_features = scaler.transform(data.drop('Город', axis=1))
```

```python
n_clusters = 6
km = KMeans(n_clusters=n_clusters)

# fit & predict clusters
data['cluster'] = km.fit_predict(scaled_features)
```

```python
data.to_excel('cluster.xlsx', index=False)
```
